```python
# Step 1: Install the moviepy library
!pip install moviepy

# Step 2: Import required libraries
from moviepy.editor import VideoFileClip
from IPython.display import Audio, display
from google.colab import files

# Step 3: Upload your own video file
print("Please upload your video file (e.g., .mp4)")
uploaded = files.upload()  # You will see a file chooser popup

# Step 4: Get the uploaded video file name
video_filename = list(uploaded.keys())[0]

# Step 5: Load the video and extract audio
video_clip = VideoFileClip(video_filename)
audio_clip = video_clip.audio

# Step 6: Save the audio to an MP3 file
audio_filename = "my_audio.mp3"
audio_clip.write_audiofile(audio_filename)

# Step 7: Play the audio inside the notebook
display(Audio(audio_filename))

# Step 8: Optional - Download the audio file
files.download(audio_filename)
```
